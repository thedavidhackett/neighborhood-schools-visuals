```python
import PyPDF2 as pypdf
reader = pypdf.PdfFileReader("charter_schools_directory.pdf")
```

```python
charter_codes = []
number_of_pages = reader.getNumPages()
```

```python
for page_num in range(1, number_of_pages - 1):
    text = reader.getPage(page_num).extractText()
    arr = text.replace("\n", "").split(" ")
    try:
        idx = arr.index('Code:')
        code = arr[idx + 1]
        charter_codes.append(code)
    except:
        continue    
```

```python
print(charter_codes)
```

```python
def create_full_code(row):
    district_str = str(row["COUNTY_DISTRICT_CODE"])
    return district_str[:3] + "-" + district_str[3:] + "-" + str(row["SCHOOL_CODE"]) 
    

stlcity_schools["FULL_CODE"] = stlcity_schools.apply(create_full_code, axis=1)
```

```python
stlcity_schools["FULL_CODE"]
```

```python
def in_charter(value):
    return value in charter_codes
stlcity_schools["CHARTER"] = stlcity_schools["FULL_CODE"].apply(in_charter)
```

```python
charter_schools = stlcity_schools[stlcity_schools["CHARTER"]]
```
